```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned and merged data
data = pd.read_csv('cleaned_merged_data.csv')

# Descriptive Statistics
print(data.describe())

```

```python

# Histogram for Salary Distribution across Departments
plt.figure(figsize=(10, 6))
sns.histplot(data[data['Department'] == 'Engineering']['Salary'], color='blue', label='Engineering', kde=True)
sns.histplot(data[data['Department'] == 'Production']['Salary'], color='red', label='Production', kde=True)
plt.title('Salary Distribution by Department')
plt.legend()
plt.show()
```

```python

# Boxplot for Operational Costs by Department
plt.figure(figsize=(10, 6))
sns.boxplot(x='Department', y='Operational Costs', data=data)
plt.title('Operational Costs by Department')
plt.show()
```

```python

# Correlation Matrix
correlation = data[['Units Produced', 'Operational Costs', 'Equipment Downtime', 'Salary']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
